```python
## Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import datetime as dt
import plotly.express as px
from capstone_functions import *
from plotly_year_functions import *

## Set options for displaying Pandas DataFrames
pd.options.display.max_columns = 50
pd.options.display.max_rows = 100
```

```python
## Import and preview age demographics dataset
age_df = pd.read_csv('Data/NC_UC_2019_demog/age_demog_2019.csv')
age_df
```

```python
error_cols = []
for col in age_df.columns:
    if 'Error' in col:
        error_cols.append(col)
        
age_df.drop(columns=error_cols, inplace=True)
age_df
```

```python
age_meta = pd.read_json('Data/NC_UC_2019_demog/age_metadata.json', typ='series')
age_meta
```

```python
col_names = []
for key in age_meta['tables']['B01001']['columns'].keys():
    col_names.append(age_meta['tables']['B01001']['columns'][key]['name'])
col_names
```

```python
age_df.columns = ['geoid', 'name'] + col_names
age_df.drop(columns=['geoid', 'Total:'], inplace=True)
age_df
```

```python
col_names = []
for col in age_df.columns:
    col = col.lower()
    col = col.replace(':', '')
    col = col.replace(' ', '_')
    col_names.append(col)
col_names
```

```python
age_df.columns = col_names
age_df.drop(index=[0,1], inplace=True)
age_df
```

```python
## Create a slice with columns corresponding to voting age males only
vote_age_males = age_df.copy()
vote_age_males = vote_age_males.iloc[:, 6:25]


## Calculate the total number of voting age males
vote_age_males['Total'] = vote_age_males.sum(axis=1)


## Group and sum Gen Z
vote_age_males['GenZ'] = vote_age_males[[
    '18_and_19_years',
    '20_years',
    '21_years', 
    '22_to_24_years'
]].sum(axis=1)


## Group and sum Millennials
vote_age_males['Millennial'] = vote_age_males[[
    '25_to_29_years',
    '30_to_34_years',
    '35_to_39_years',
    '40_to_44_years'
]].sum(axis=1)


## Group and sum Generation X
vote_age_males['GenX'] = vote_age_males[[
    '45_to_49_years',
    '50_to_54_years'
]].sum(axis=1)


## Group and sum Baby Boomers
vote_age_males['Boomer'] = vote_age_males[[
    '55_to_59_years',
    '60_and_61_years',
    '62_to_64_years',
    '65_and_66_years',
    '67_to_69_years',
    '70_to_74_years'
]].sum(axis=1)


## Group and sum Greatest and Silent generations
vote_age_males['Greatest-Silent'] = vote_age_males[[
    '75_to_79_years',
    '80_to_84_years',
    '85_years_and_over'
]].sum(axis=1)


## Add gender_code column
vote_age_males['gender_code'] = 'M'

vote_age_males = vote_age_males[[
    'gender_code',
    'GenZ',
    'Millennial',
    'GenX',
    'Boomer',
    'Greatest-Silent',
    'Total'
]]
vote_age_males
```

```python
## Create a slice with columns corresponding to voting age females only
vote_age_females = age_df.copy()
vote_age_females = vote_age_females.iloc[:, 30:]

## Calculate the total number of voting age females
vote_age_females['Total'] = vote_age_females.sum(axis=1)


## Group and sum Gen Z
vote_age_females['GenZ'] = vote_age_females[[
    '18_and_19_years',
    '20_years',
    '21_years', 
    '22_to_24_years'
]].sum(axis=1)


## Group and sum Millennials
vote_age_females['Millennial'] = vote_age_females[[
    '25_to_29_years',
    '30_to_34_years',
    '35_to_39_years',
    '40_to_44_years'
]].sum(axis=1)


## Group and sum Generation X
vote_age_females['GenX'] = vote_age_females[[
    '45_to_49_years',
    '50_to_54_years'
]].sum(axis=1)


## Group and sum Baby Boomers
vote_age_females['Boomer'] = vote_age_females[[
    '55_to_59_years',
    '60_and_61_years',
    '62_to_64_years',
    '65_and_66_years',
    '67_to_69_years',
    '70_to_74_years'
]].sum(axis=1)


## Group and sum Greatest and Silent generations
vote_age_females['Greatest-Silent'] = vote_age_females[[
    '75_to_79_years',
    '80_to_84_years',
    '85_years_and_over'
]].sum(axis=1)


## Add gender_code column
vote_age_females['gender_code'] = 'F'

vote_age_females = vote_age_females[[
    'gender_code',
    'GenZ',
    'Millennial',
    'GenX',
    'Boomer',
    'Greatest-Silent',
    'Total'
]]
vote_age_females
```

```python
## Concatenate both gender dataframes
age_gender_df = pd.concat([vote_age_females, vote_age_males],
                          ignore_index=True)
age_gender_df.set_index('gender_code', inplace=True)

## Add a row that sums the total number of voters in each group
age_gender_df.loc['All'] = age_gender_df.sum(axis=0)
age_gender_df
```
