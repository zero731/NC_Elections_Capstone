```python
## Import and preview dataset
gen_elecs_df = pd.read_csv('Data/UC_gen_elecs.gz')

# Recast target variable as binary 0 or 1
gen_elecs_df['vote_bin'] = np.where(gen_elecs_df['vote_bin']=='Y', 1, 0)
# Recast drivers_lic variable as binary 'Y'/'N'
gen_elecs_df['drivers_lic'] = np.where(gen_elecs_df['vote_bin']=='License', 'Y', 'N')

display(gen_elecs_df.head())

## Check df metadata
gen_elecs_df.info()
```

# MODEL

```python
## Import necessary functions/methods/classes for modeling
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import classification_report, plot_confusion_matrix, plot_roc_curve

## Import classes necessary for building preprocessing pipelines
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler, OneHotEncoder, LabelEncoder
from sklearn.compose import ColumnTransformer
```

## Prepare DataFrame for Modeling - Train Test Split and Preprocessing

```python
## Create new DataFrame from the full cleaned df that only consists of 
  ## the target ('vote_bin') and desired predictors
model_all_df = gen_elecs_df.copy()
model_all_df = model_all_df[['vote_bin', 'gen_grp', 'party_grp',
                     'race_grp', 'gender_code', 'birth_reg_other',
                     'drivers_lic', 'city_grp']]

## With binary voted vs. did not vote target, can keep even those voters 
  ## with provisional or transfer ballots because they did go vote (1).
    ## Thus, no need to drop the 'Other' voting method

## Inspect the resulting df
display(model_all_df.head())
len(model_all_df)
```

```python
## Check class imbalance
check_unique('vote_bin', model_all_df)
```

```python
## Define target variable
target = 'vote_bin'

## Separate features (X) and target (y) for train-test-split
X_all = model_all_df.drop(columns=target, axis=1).copy()
y_all = model_all_df[target].copy()
```

```python
## Define random seed to use for train-test-split and 
 ## classifiers for reproducibility
random_seed = 319

## Split the data into training and test sets prior to preprocessing
X_all_train, X_all_test, y_all_train, y_all_test = train_test_split(
    X_all, y_all, test_size=0.25, random_state=random_seed
)
```

```python
## Check for class imbalance across all sets of y
print('**original**\n', y_all.value_counts(normalize=True),
      '\n------\n')
print('**y_all_train**\n', y_all_train.value_counts(normalize=True),
      '\n------\n')
print('**y_all_test**\n', y_all_test.value_counts(normalize=True),
      '\n------\n')
```

```python
## Make a list of all columns that are currently object dtype
cat_cols = list(X_all_train.select_dtypes('O').columns)
cat_cols
```

```python
## Create a pipeline for one hot encoding categorical columns
cat_transformer = Pipeline(steps = [
    ('ohe', OneHotEncoder(handle_unknown='error', 
                          sparse=False,
                          drop='if_binary'))])
cat_transformer
```

```python
## Combine num and cat transformer pipelines into a single pipeline
  ## for preprocessing X
preprocessing = ColumnTransformer(transformers=[
    ('cat', cat_transformer, cat_cols)
])
preprocessing
```

```python
## Preprocess training and test predictors (X)
X_all_train_tf = preprocessing.fit_transform(X_all_train)
X_all_test_tf = preprocessing.transform(X_all_test)
```

```python
## Get the feature names in the order they appear in preprocessed data
feature_names_all = preprocessing.named_transformers_['cat'].named_steps['ohe'].get_feature_names(cat_cols)
feature_names_all
```

## Random Forest

### Baseline Random Forest

```python
## Use preprocessor as the first step in modeling Pipeline 
rf_all = Pipeline(steps=[('preprocessing', preprocessing),
                          ('rf', 
                           RandomForestClassifier(random_state=random_seed))])

rf_all.fit(X_all_train, y_all_train)
```

```python
rf_all.named_steps['rf'].classes_
```

```python
eval_bin_clf(rf_all, X_all_test, y_all_test,
                model_descr='Baseline All Year Random Forest with Default Params');
```

### Tuning Random Forest Hyperparameters

```python
## Define a grid of different hyperparameter values to be tested
params = {
    'rf__criterion':['gini', 'entropy'],
    'rf__max_depth':[4, 5, 6],
    'rf__max_features': [0.3, 0.4, 0.5],
    'rf__class_weight':[None, 'balanced']
}

## run the grid search to test parameters and find the best estimator 
rf_all_bacc_grid = fit_grid_clf(rf_all, params, 
                                X_all_train, y_all_train,
                                X_all_test, y_all_test,
                                model_descr='Best All Year RF Tuned for Balanced Accuracy',
                                score='balanced_accuracy',
                                bin_target=True)
```

### Random Forest with SMOTE

```python
import six
import sys
sys.modules['sklearn.externals.six'] = six
from imblearn.over_sampling import SMOTE
from imblearn.pipeline import Pipeline as imbPipeline
```

```python
## Use preprocessor as the first step in modeling Pipeline 
rf_all_smote = imbPipeline(steps=[
    ('preprocessing', preprocessing),
    ('sampler', SMOTE(random_state=random_seed, n_jobs=-1)),
    ('rf', RandomForestClassifier(random_state=random_seed))])

rf_all_smote.fit(X_all_train, y_all_train)
```

```python
eval_bin_clf(rf_all_smote, X_all_test, y_all_test,
             model_descr='SMOTE All Year Random Forest with Default Params');
```

### Tuning Random Forest with SMOTE Hyperparameters

```python
## Define a grid of different hyperparameter values to be tested
params = {
    'sampler__k_neighbors':[3, 5],
    'rf__criterion':['gini'],
    'rf__max_depth':[5, 6],
    'rf__max_features': [0.3, 0.4, 0.5]
}

## run the grid search to test parameters and find the best estimator 
rf_all_smote_bacc_grid = fit_grid_clf(rf_all_smote, params, 
                                      X_all_train, y_all_train,
                                      X_all_test, y_all_test,
                                      model_descr='Best All Year SMOTE RF Tuned for Balanced Accuracy',
                                      score='balanced_accuracy',
                                      bin_target=True)
```

```python
plot_feat_importance(rf_all_smote_bacc_grid.best_estimator_,
                     'rf', feature_names_all,
                     model_title='Best All Year SMOTE RF Tuned for Balanced Accuracy');
```

# INTERPRET

```python
## Import necessary libraries
import xgboost
import shap

## Load JS visualization code to notebook
shap.initjs()
```

```python
## Convert transformed (preprocessed) X train set into pandas DataFrame
X_all_train_df = pd.DataFrame(X_all_train_tf, columns=feature_names_all)
X_all_train_df.head()
```

```python
## Store the booster (trained model)
all_model = xgb_all_bacc_grid.best_estimator_.named_steps['xgb'].get_booster()

## Explain the model's predictions using SHAP
all_explainer = shap.TreeExplainer(all_model)
all_shap_values = all_explainer.shap_values(X_all_train_df)
```

```python
## Print model prediction, true label, and shap force plot for seventh row
  ## of training set
shap_force(xgb_all_bacc_grid.best_estimator_, 'xgb',
           X_all_train_df, y_all_train,
           6, all_explainer, all_shap_values)
```

The plot above visualizes how the model is making its prediction for the seventh individual in the training set. 
> The seventh individual did not vote and the model predicted this correctly. The model was pushed towards the prediction of "no vote" by the fact that the individual is a Millennial. The fact that the individual was male and not white also pushed the model towards the correct prediction for this individual. The facts that he does not live in Monroe pushed the model in the opposite direction, but not enough to overwhelm the factors more predictive of "no vote" (0).

```python
## Print model prediction, true label, and shap force plot for third row
  ## of training set
shap_force(xgb_all_bacc_grid.best_estimator_, 'xgb',
           X_all_train_df, y_all_train,
           2, all_explainer, all_shap_values)
```

The plot immediately above visualizes how the model is making its prediction for the third individual in the training set. 
> The third individual voted and the model predicted this correctly. The model was pushed towards the prediction of "vote" (1) by the fact that the individual is not a Millennial (they belong to the Baby Boomer generation), and they registered as a Republican. The fact the this person's birth region information is not missing pushed the model in the incorrect direction towards "no vote" (0), but not enough to overwhelm the factors more predictive of "vote" (1).

The plot below shows the impact each feature had on the model's output (predictions) quantified in terms of SHAP values. The top 4 features (after which there is a drop in impact) are `gen_grp_Millennial`, `birth_reg_other_Missing`, `party_grp_Rep`, and `gen_grp_GenZ`. Obviously age group is important because it makes up 2 of the top 4 features. Whether or not a person's voter registration record lists their birthplace is the second most impactful feature on model output, and whether or not the voter registered as a Republican is the third most impactful feature. Whether or not the person belongs to Gen Z was the 4th most impactful predictor.

```python
## Display mean abs(SHAP value) for each feature for the full training set
shap.summary_plot(all_shap_values, X_all_train_df, plot_type='bar')
```

The plot below is similar to that above in that it ranks the top features based on mean SHAP values (so the ordering is the same). Each point represents a voter in the training set. The grey vertical line at 0.0 represents SHAP values that had 0 impact on model output, so the farther the points are from the grey line, the more important the feature was in terms of predicting that voter's participation in the election (whether they voted or not). Negative SHAP values to the left of the grey line push the model towards a prediction of 0 or "No Vote" whereas positive SHAP values push the model towards a prediction of 1 or "Vote". 

Since all the features were label encoded, each feature is a binary variable representing whether the voter does or does not belong to the category listed at the end of the feature name (so 1 for `gen_grp_Millenial` means the person is a millennial). Thus high values of 1 are color coded red and low values of 0 are color coded blue.

**Top 7 Features Explained**:

1. `gen_grp_Millennial` - Being a millenial (1, red) pushed the model towards a prediction of "No Vote".


2. `birth_reg_other_Missing` - Voters whose record was missing their birthplace information (1, red) were more likely to vote.


3. `party_grp_Rep` - Being a Republican (1, red) pushed the model towards a prediction of "Vote".


4. `gen_grp_GenZ` - Belonging to the Gen Z (1, red) pushed the model towards a prediction of "No Vote".
